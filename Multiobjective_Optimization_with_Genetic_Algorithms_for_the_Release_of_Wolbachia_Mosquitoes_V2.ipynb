# **Conceptualization of the proposal**

Definition

> *Hybrid Control Strategies:* We define hybrid control strategies as those strategies that use suppression and replacement control.
____________________________________
 Thus, we formulated the following problem: At what time intervals and density can male and female mosquitoes infected with Wolbachia be released to ensure optimal intervention?
 ___________________________________

# **Dynamics with aquatic and aerial phase**

Introducing input variables $v_{L\gamma}, v_{A\gamma}$ corresponding to releases of infected larvae or adults male or famele, yields the following *controlled* system:

\begin{align}
\nonumber
\dot{L}^\eta_\gamma = & r^\eta \alpha^\eta_\gamma(A) - \hat{\mu}^\eta_\gamma(|L|) L^\eta_\gamma - \nu L^\eta_\gamma +\delta_\eta^W v_{Li},
\quad \eta=U,W,\ \gamma=M,F\\
\dot{A}^\eta_\gamma = & \nu L^\eta_\gamma - \mu^\eta_\gamma(|A|) A^\eta_\gamma + \delta_\eta^W v_{Ai},
\qquad \eta=U,W,\ \gamma=M,F
\end{align}


# **Slow reproductive phase population dynamic**

 In the same way that the system  **S** was obtained,  one consider instead _the algebro-differential model_:

\begin{equation}
		L^\eta_\gamma = \frac{r^\eta \alpha^\eta_\gamma(A)}{\hat{\mu}^\eta_\gamma(b^*(\alpha(A))) + \nu}, \quad
		\dot{A}^\eta_\gamma = \nu L^\eta_\gamma - \mu^\eta_\gamma A^\eta_\gamma + \delta_\eta^W v_{A\gamma}
\end{equation}
 where  $ b:=v^T L $, such that the first eq. is the solution of $ \alpha^\eta_{\gamma}(A)-\mu_{\gamma}(v^T L)L_{\gamma}-\nu_{\gamma}L_{\gamma}=0$.

Defining for any  $b>0$ and $m_{\gamma}^\eta(b):=\frac{\nu r^\eta}{\hat{\mu}_{\gamma}^\eta(b)+\nu}$, one obtains the model:
\begin{equation}
\dot A^\eta_\gamma = m^\eta_\gamma (b^*(\alpha(A)))\, \alpha^\eta_\gamma (A) - \mu^\eta_\gamma(|A|) A^\eta_\gamma + \delta_\eta^W v_{A\gamma},
\quad \eta=U,W,\ \gamma=M,F
\end{equation}
where the functions $ m^\eta_\gamma $ are _decreasing_.

# **General hypotheses**

We make the following general hypotheses to introduce the dynamical system under study.

___
For any $\eta=U,W$, for any $\gamma=M,F$, for simplicity, we assume in the sequel that the previous inequalities are _strict_ in the following sense:

- either $\hat{\mu}^U_M < \hat{\mu}^W_M$ or $\mu^U_F < \mu^W_F$.

___

Furthermore, $ b^*(\alpha(A)) \approx |A| $, such that for simplicity $ m^\eta_\gamma (b^*(\alpha(A)))  := m^\eta_\gamma (|A|)$, where the functions $ m^\eta_\gamma $ are _decreasing_.

___

**The model: Wolbachia$\cap$Sex inheritance**


\begin{equation}
\nonumber
 A=( A_M^U \ \  A_F^U \ \   A_M^W \ \ A_F^W )^T; \ \ |A|:=1_4^T A;  \ \ A_{\gamma}=( A_{\gamma}^U \ \  A_{\gamma}^W)^T, \ \  \gamma = M,F.
\end{equation}


Introducing input variables $v_{\gamma}$ corresponding to releases of infected adults M or F, yields the following _controlled_ system:


\begin{align}
\nonumber
\dot{A}^\eta_{\gamma} = &  \alpha^\eta_{\gamma}(A)m^{\eta}_{\gamma}(|A|) - \hat{\mu}^\eta_{\gamma}(|A|) A^\eta_{\gamma} +\delta_\eta^W v_{\gamma},
\quad \eta=U,W,\ \gamma=M,F
\end{align}


\begin{equation}
\nonumber
\alpha^\eta_{\gamma}\ :\ \mathbb{R}_+^4\setminus\{0_4\}\to\mathbb{R}_+, \quad \alpha^\eta_{\gamma}(A_M,A_F) := \frac{1}{|A|} A_{\gamma'}^T G^\eta A_{\gamma}, \ \ \gamma' \neq \gamma
\end{equation}


\begin{equation}
\nonumber
G^U = \begin{pmatrix}
1 & 0 \\ 1-\sigma & 0
\end{pmatrix},\quad
G^W = \begin{pmatrix}
0 & 1 \\ 0 & 1
\end{pmatrix}, \quad
\delta_{\eta}^{\eta'} :=
    \begin{cases}
            1, &         \text{if } \eta=\eta',\\
            0, &         \text{if } \eta\neq \eta'
    \end{cases}
\end{equation}

## **The proposed model**


The controlled system is written compactly under vector form as


\begin{equation}
\nonumber
\dot A = \overbrace{m(|A|)\, \alpha(A)}^{\text{recruitment}} - \underbrace{\mu(|A|) A}_{\text{mortality}}+ \overbrace{V}^{\text{input variables}},
\end{equation}


\begin{equation}
\nonumber
m(|A|):=\diag\{ m^\eta_\gamma (|A|) \},\ \mu(|A|) :=\diag\{\mu^\eta_\gamma(|A|)\}, \ \ V = (0,0,v_M,v_F)
\end{equation}


\begin{equation}
\nonumber
\alpha(A_M,A_F)\ :\ \mathbb{R}_+^4\setminus\{0_4\}\to\mathbb{R}_+^4, \ \ \overbrace{\alpha(A_M,A_F) := \sum_{\eta=U,W}\sum_{\gamma=M,F} \alpha^\eta_\gamma(A) (e^\eta\otimes e_\gamma)}^{\text{Inheritance operator}}.
\end{equation}


where $ e^U=(1 \ \ 0)^T $,  $ e^W=(0 \ \ 1)^T $ and $ e_M=(1 \ \ 0 )^T $, $ e_F=(0 \ \ 1)^T $.

For the seasonal effect of temperature, the simple harmonic equation is expressed as:

$s(t) = A+B cos(2πt/365)+C sen(2πt/365)$,

such that the annual phases are given by $2πt/365$. The parameter $A$ is the regression constant or mesor, which is equivalent to the average value of the function $s(t)$; $B$ is the coefficient of the dependent variable cosine; C is the coefficient of the dependent variable sine. The parameters$\{A,B,C\}$they are unknown parameters estimated based on climatic temperature data for a given location.

 For climate data for Paraguay from the 1960s to the present, these are:
 $A = 22.6071877371859$

 $B = 4.67698314471263$

 $C = 1.08897678248751$

 Depending on the location considered, these values ​​may change.

```r
## ================================================================================
## Hybrid Wolbachia Introgression Suppression-Replacement Control Models
 ===========================================================================================

## Note:
## The function w(t) is an argument (input) to the model
## Models the release of infected male or female mosquitoes
# https://www.rdocumentation.org/packages/deSolve/versions/1.40/topics/ode

TSW_Model = function(times, x, parms,input)  {
  with(as.list(c(parms, x)), {
       s = input(times)
       #  the time-dependent oviposition rate mUM0_t, mUF0_t, mWM0_t, mWF0_t
       mUM0_t = mUM0_c*s*(s-mUM0_T0)*(mUM0_Tm-s)^(1/2)
       mUF0_t = mUF0_c*s*(s-mUF0_T0)*(mUF0_Tm-s)^(1/2)
       mWM0_t = (1-m_w)*mWM0_c*s*(s-mWM0_T0)*(mWM0_Tm-s)^(1/2)
       mWF0_t = (1-m_w)*mWF0_c*s*(s-mWF0_T0)*(mWF0_Tm-s)^(1/2)
       # time-dependent death rates of adult mosquitoes muUM0_t, muUF0_t, muWM0_t, muWF0_t
       muUM0_t = (muUM0_c*(s-muUM0_T0)*(muUM0_Tm-s))^(-1)
       muUF0_t = (muUF0_c*(s-muUF0_T0)*(muUF0_Tm-s))^(-1)
       muWM0_t = (1+mu_w)*(muWM0_c*(s-muWM0_T0)*(muWM0_Tm-s))^(-1)
       muWF0_t = (1+mu_w)*(muWF0_c*(s-muWF0_T0)*(muWF0_Tm-s))^(-1)
       # Import
       WM_t = WM*(times >= lowerM & times <= upperM)  # Infected males
       WF_t = WF*(times >= lowerF & times <= upperF) # Infected females
    dx_UM = ((x_UF*(x_UM + (1-sigma)*x_WM))/(x_UM+x_UF+x_WM+x_WF))*(mUM0_t - mUM*(x_UM+x_UF+x_WM+x_WF))-(x_UM*(muUM0_t + muUM*(x_UM+x_UF+x_WM+x_WF))) # Machos no infectados
    dx_UF = ((x_UF*(x_UM + (1-sigma)*x_WM))/(x_UM+x_UF+x_WM+x_WF))*(mUF0_t - mUF*(x_UM+x_UF+x_WM+x_WF))-(x_UF*(muUF0_t + muUF*(x_UM+x_UF+x_WM+x_WF))) # Hembras no infectadas
    dx_WM = ((x_WF*(x_UM + x_WM))/(x_UM+x_UF+x_WM+x_WF))*(mWM0_t - mWM*(x_UM+x_UF+x_WM+x_WF))-(x_WM*(muWM0_t + muWM*(x_UM+x_UF+x_WM+x_WF))) + WM_t # Machos infectados
    dx_WF = ((x_WF*(x_UM + x_WM))/(x_UM+x_UF+x_WM+x_WF))*(mWF0_t - mWF*(x_UM+x_UF+x_WM+x_WF))-(x_WF*(muWF0_t + muWF*(x_UM+x_UF+x_WM+x_WF))) + WF_t # Hembras infectadas
    res <- c(dx_UM, dx_UF, dx_WM, dx_WF)
    return(list(c(dx_UM, dx_UF, dx_WM, dx_WF),
    s = s,
    WM_t = WM_t,
    WF_t = WF_t,
    mUM0_t = mUM0_t,
    mUF0_t = mUF0_t,
    mWM0_t = mWM0_t,
    mWF0_t = mWF0_t,
    muUM0_t = muUM0_t,
    muUF0_t = muUF0_t,
    muWM0_t = muWM0_t,
    muWF0_t = muWF0_t))
  })
}

## The parameters
parms = c(
  sigma = 1,
  #Cytoplasmic incompatibility coefficient. Represents the reduction in successful mating between infected males and uninfected females, leading to fewer viable offspring.
  #recruitment reduction ratio due to infection
  m_w = 0.20,# Recruitment reduction ratio due to infection. The proportion by which infected mosquito recruitment (birth rate) is reduced compared to uninfected mosquitoes.
  # proportion of increased mortality due to infection
  mu_w = 0.20,#Proportion of increased mortality due to infection. The additional mortality rate imposed on infected mosquitoes compared to uninfected ones.
  #  the time-dependent oviposition rate parameters
  mUM0_c = 0.0058,
  mUF0_c = 0.0058,
  mWM0_c = 0.0058,
  mWF0_c = 0.0058,
  #Baseline oviposition rate scaling constants for uninfected (U) and Wolbachia-infected (W) males (M) and females (F).These are reference values for egg-laying rates at optimal temperatures.
  mUM0_T0 = 14.0343,
  mUF0_T0 = 14.0343,
  mWM0_T0 = 14.0343,
  mWF0_T0 = 14.0343,
  #Minimum temperature thresholds for oviposition. Below this temperature, egg-laying rates drop significantly.
  mUM0_Tm = 39.0899,
  mUF0_Tm = 39.0899,
  mWM0_Tm = 39.0899,
  mWF0_Tm = 39.0899,
  #Maximum temperature thresholds for oviposition. Above this temperature, egg-laying rates decline sharply.
  mUM = 0.0001,
  mUF = 0.0001,
  mWM = 0.0002,
  mWF = 0.0002,
  #Additional mortality or stress factors affecting oviposition.These may represent residual effects (e.g., due to extreme temperatures or other stressors).
  # time-dependent death rates of adult mosquitoes parameters
  muUM0_c = 0.1037,
  muUF0_c = 0.1037,
  muWM0_c = 0.1037,
  muWF0_c = 0.1037,
  #Baseline mortality rate scaling constants for uninfected (U) and Wolbachia-infected (W) males (M) and females (F).Reference mortality rates at optimal temperatures.
  muUM0_T0 = 6.4429,
  muUF0_T0 = 6.4429,
  muWM0_T0 = 6.4429,
  muWF0_T0 = 6.4429,
  #Minimum temperature thresholds for mortality. Below this temperature, mortality rates increase sharply.
  muUM0_Tm = 41.4382,
  muUF0_Tm = 41.4382,
  muWM0_Tm = 41.4382,
  muWF0_Tm = 41.4382,
  # Maximum temperature thresholds for mortality. Above this temperature, mortality rates rise significantly.
  muUM = 0.0002,
  muUF = 0.0002,
  muWM = 0.0003,
  muWF = 0.0003,
  #Additional mortality factors (e.g., due to environmental stress or infection).These may account for extra mortality beyond temperature effects.
  WM = 4000* 0.05000000,# Initial number of Wolbachia-infected male mosquitoes in the population.
  WF = 4000*0.05023465 ,# Initial number of Wolbachia-infected female mosquitoes in the population.
  lowerM = 146, # Minimum age (in days) for male mosquitoes to be included in a specific life stage
  upperM = 146+44,# Maximum age (in days) for male mosquitoes in the same life stage.
  lowerF = 154,# Minimum age (in days) for female mosquitoes to be included in a life stage
  upperF = 154+48 #Maximum age (in days) for females in the same stage.
)

## vector of timesteps
times = seq(0, 365*3)

## external signal with rectangle impulse
signal = data.frame(times = times,
                     s = rep(NA, length(times)))

 A = 22.6071877371859

 B = 4.67698314471263

 C = 1.08897678248751

signal$s = A + B*cos(2*pi*signal$times/365) + C*sin(2*pi*signal$times/365)

sigimp <- approxfun(signal$times, signal$s, rule = 2)

## Start values for steady state
xstart <- c(
  x_UM = 4000,
  x_UF= 4000,
  x_WM= 0,
  x_WF = 0)

## Solve model
out <- ode(y = xstart, times = times,
           func = TSW_Model, parms = parms, , input = sigimp)

## Default plot method
plot(out)
```

```r
# 1.Generate the chart with custom scale
plot <- as.data.frame(out) %>%
  pivot_longer(!time, names_to = "var", values_to = "size") %>%
  subset(var %in% c("x_UF", "x_UM", "x_WF", "x_WM", "WF_t", "WM_t")) %>%
  ggplot(aes(x = time, y = size, color = var)) +
  geom_line(linewidth = 0.8) +
  theme_minimal() +
  theme(
    legend.position = 'none',
    text = element_text(size = 12),
    axis.title = element_text(size = 14, ),
    axis.text = element_text(size = 12),
    strip.text = element_text(size = 12, )
  ) +
  facet_wrap(~var, scales = "free_y", ncol = 1) +
  labs(x = "Time", y = "Size") +
  scale_y_continuous(
    limits = c(0, NA),  # Force start at 0
    breaks = function(x) c(0, max(x)),  # It only shows 0 and the maximum
    labels = function(x) ifelse(x == 0, "0", round(x, 1))  # Format: 0 and maximum with 1 decimal
  )

# Show the graph
print(plot)

# 2. Save the chart
ggsave("mosquito_plot_simple.png", plot,
       width = 4, height = 3.2, dpi = 300, bg = "white")
```
